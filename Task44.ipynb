В ячейке ниже представлен код генерирующий <u>_DataFrame_</u>, которая состоит всего из 1 столбца. Ваша задача перевести его в <u>_one hot_</u> вид. Сможете ли вы это сделать без <u>_get_dummies_</u>?
```python
import random

data.head()
```


````python
import pandas as pd
import random 
lst = ['robot'] * 10  
lst += ['human'] * 10  
random.shuffle(lst)  
data = pd.DataFrame({'whoAmI':lst})
data.head(10)
````

````python
# Метод get_dummies
pd.get_dummies(data['whoAmI']).head(10)
````

````python
# Без метода get_dummies
pd.DataFrame({'human': data['whoAmI'] == 'human',
              'robot': data['whoAmI'] == 'robot'}).astype(int).head(10)
````
